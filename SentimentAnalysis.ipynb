```python
!pip install textblob

import pandas as pd
import numpy as np
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!pip install textblob

# Simulated feedback dataset
data = {
    'Candidate_ID': range(1, 11),
    'Feedback': [
        "The process was smooth and easy, thank you!",
        "Very disappointed with the delay in processing.",
        "Staff was helpful and responded quickly.",
        "I had a terrible experience. No one answered my calls.",
        "Happy with the guidance I received.",
        "Service was average, not very responsive.",
        "Great support from the documentation team.",
        "Unsatisfactory experience overall.",
        "The consultants were friendly and knowledgeable.",
        "Iâ€™m not satisfied with the way my case was handled."
    ]
}

df = pd.DataFrame(data)
df.head()
```

```python
# Apply TextBlob sentiment analysis
def analyze_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity  # -1 to 1
    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['Feedback'].apply(analyze_sentiment)
df.head()
```

```python
# Plot sentiment distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Sentiment', palette='Set2')
plt.title("Candidate Feedback Sentiment Distribution")
plt.ylabel("Number of Responses")
plt.show()
```

```python
# Exporting results to CSV (optional)
df.to_csv("sentiment_feedback_results.csv", index=False)
```
